```python
import pandas as pd
import csv

head = ['symboling', 'normalized_losses', 'make', 'fuel_type', 'aspiration', 'num_doors', 'body_style', 'drive_wheels', 'engine_location', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_type', 'num_cylinders', 'engine_size', 'fuel_system', 'bore', 'stroke', 'compression_ratio', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']

cars = pd.read_csv('imports-85.data', names=head)

cars.columns
```

```python
numeric_col = ['normalized_losses', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 
       'bore', 'stroke',
       'compression_ratio', 'horsepower', 'peak_rpm', 'city_mpg',
       'highway_mpg', 'price']

numeric_car = cars[numeric_col]

numeric_car.shape
```

```python
import numpy as np

numeric_car = numeric_car.replace("?", np.nan)
numeric_car
```

```python
numeric_car = numeric_car.astype('float')
numeric_car.isnull().sum()
```

```python
numeric_car = numeric_car.dropna(subset=['price'])
numeric_car.isnull().sum()

# Always clean up the target column first
```

```python
numeric_car = numeric_car.fillna(numeric_car.mean())
numeric_car.isnull().sum()

# Replace NaN with column mean
```

```python
price_col = numeric_car['price']
numeric_car = (numeric_car - numeric_car.min()) / (numeric_car.max() - numeric_car.min())
numeric_car['price'] = price_col
numeric_car.head(5)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(train_col, target_col, df):
    knn = KNeighborsRegressor()
    np.random.seed(1)
    
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    limit = int(len(rand_df) / 2 )
    df_train = rand_df.iloc[0:limit]
    df_test = rand_df.iloc[limit:]
    knn.fit(df_train[[train_col]], df_train[target_col])
    predict = knn.predict(df_test[[train_col]])
    rmse = mean_squared_error(df_test[target_col], predict) ** (1/2)
    return rmse

rmse_dict = {}

for each in numeric_car.columns.drop('price'):
    rmse_each = knn_train_test(each, 'price', numeric_car)
    rmse_dict[each] = rmse_each
    
rmse_series = pd.Series(rmse_dict)
rmse_series.sort_values()
```

```python
def knn_train_test_k(train_col, target_col, df, k):
    knn = KNeighborsRegressor(n_neighbors=k)
    np.random.seed(1)
    
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    limit = int(len(rand_df) / 2 )
    df_train = rand_df.iloc[0:limit]
    df_test = rand_df.iloc[limit:]
        
    knn.fit(df_train[[train_col]], df_train[target_col])
    predict = knn.predict(df_test[[train_col]])
    rmse = mean_squared_error(df_test[target_col], predict) ** (1/2)
    return rmse

k_values = [1, 3, 5, 7, 9]


rmse_all_k = {}

for col in numeric_car.columns.drop('price'):  
    rmse_k = {}
    for i in k_values:
        rmse_i = knn_train_test_k(col, 'price', numeric_car, i)
        rmse_k[i] = rmse_i
    rmse_all_k[col] = rmse_k
    
rmse_all_k
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for k, v in rmse_all_k.items():
    y = list(v.values())
    plt.plot(k_values, y)
    plt.xlabel("k values")
    plt.ylabel('RMSE')
    
plt.show()
```

```python
feature_avg_rmse = {}
for k,v in rmse_all_k.items():
    avg_rmse = np.mean(list(v.values()))
    feature_avg_rmse[k] = avg_rmse
series_avg_rmse = pd.Series(feature_avg_rmse)
series_avg_rmse.sort_values()
```

```python
def knn_trainlist_test(train_col, target_col, df):
    knn = KNeighborsRegressor()
    np.random.seed(1)
    
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    limit = int(len(rand_df) / 2 )
    df_train = rand_df.iloc[0:limit]
    df_test = rand_df.iloc[limit:]
        
    knn.fit(df_train[train_col], df_train[target_col])
    predict = knn.predict(df_test[train_col])
    rmse = mean_squared_error(df_test[target_col], predict) ** (1/2)
    return rmse

best_feature = ['horsepower', 'width', 'curb_weight', 'highway_mpg', 'length']

multi_rmse = {}

for i in range(2, 6):
    train_list = best_feature[:(i-1)]
    rmse_each = knn_trainlist_test(train_list, 'price', numeric_car)
    multi_rmse[i] = rmse_each
    
multi_rmse
```

```python
from collections import OrderedDict

def knn_trainlist_test_k(train_col, target_col, df, k):
    knn = KNeighborsRegressor(n_neighbors = k)
    np.random.seed(1)
    
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    limit = int(len(rand_df) / 2 )
    df_train = rand_df.iloc[0:limit]
    df_test = rand_df.iloc[limit:]
        
    knn.fit(df_train[train_col], df_train[target_col])
    predict = knn.predict(df_test[train_col])
    rmse = mean_squared_error(df_test[target_col], predict) ** (1/2)
    return rmse

rmse25 = {}

for i in range(3, 6):
    train_list = best_feature[:(i-1)]
    name = str(i) + " best features"
    k_list = OrderedDict()
    for k in range(1, 26):
        rmse_each = knn_trainlist_test_k(train_list, 'price', numeric_car, k)
        k_list[k] = rmse_each
    rmse25[name] = k_list
    
rmse25
```

```python
for k, v in rmse25.items():
    y = list(v.values())
    plt.plot(range(1, 26), y, label=k)
    plt.xlabel("k values")
    plt.ylabel('RMSE')
    
plt.legend(loc='upper left', bbox_to_anchor=(0.5, 0.5))    
plt.show()
```

With k less than 5, the three model differes in RMSE, with 5 best feature model appears to have the least error. However, as k increases, the difference is diminished until all models have the same RMSE.
